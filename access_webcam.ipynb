## Import dependencies

```python
import cv2
from matplotlib import pyplot as plt
```

## Conect to Webcam

```python
cap = cv2.VideoCapture(0)
```

```python
# get a frame from device
ret, frame = cap.read()
```

```python
plt.imshow(frame)
cap.release()
```

```python
def take_photo():
    cap = cv2.VideoCapture(0)
    ret, frame = cap.read()
    cv2.imwrite('webcamphoto.jpg', frame)
    plt.imshow(frame)
    cap.release()
```

```python
take_photo()
```

## Render in Real Time

```python
# Connect to cam
cap = cv2.VideoCapture(0)

while cap.isOpened():
    ret, frame = cap.read()
    
    # Show image 
    cv2.imshow('Webcam', frame)
    
    # check if any keys are being pressed, if this is 'q' break out of while loop and release webcam & close frame
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
